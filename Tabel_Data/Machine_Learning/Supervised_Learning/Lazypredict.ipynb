## Classifier

```python
import lazypredict

# 分類
from lazypredict.Supervised import LazyClassifier

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

データセットの用意

```python
iris_dataset = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], test_size=0.25,  random_state=0)
```

モデルの定義と学習を行う

```python
clf = LazyClassifier(verbose=1,ignore_warnings=True,predictions=True)
models, predictions = clf.fit(X_train, X_test, y_train, y_test)
```

計算したモデルの結果の一覧を表示

```python
models
```

モデルごとの予測値

```python
predictions.head()
```

## Regressor

```python
# 回帰
from lazypredict.Supervised import LazyRegressor
from sklearn.datasets import load_boston
```

データセットの用意

```python
boston_dataset = load_boston()
X_train, X_test, y_train, y_test = train_test_split(boston_dataset['data'], boston_dataset['target'], test_size=0.25,  random_state=0)
```

回帰でも，分類と同様にモデルの定義と予測を行う

```python
reg = LazyRegressor(verbose=1,ignore_warnings=True,predictions=True)
models, predictions = reg.fit(X_train, X_test, y_train, y_test)
```

計算したモデルの結果の一覧

```python
models
```
