# 主成分分析(PCA)

```python
from sklearn.decomposition import PCA #主成分分析器

import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
import pandas as pd
```

## データの用意

```python
dataset = load_wine()
```

```python
dataset.data
```

```python
label = dataset.target
```

## PCA

```python
pca = PCA(n_components=2)
x_pca = pca.fit_transform(dataset.data)
```

```python
plt.scatter(x_pca[:,0],x_pca[:,1], c=label)
plt.xlabel("tsne-1")
plt.ylabel("tsne-2")
```

## 寄与率
* 各主成分の重要性を表す
* 各主成分によってデータ全体の説明できる割合を表している(合計すると1になる)

```python
#累積寄与率の表示
pca.explained_variance_ratio_
```

## 因子負荷量
* 各変数の各主成分への影響力
* 各主成分の意味の推定もできる
* 「行」が主成分、「列」が変数

```python
pca.components_
```
