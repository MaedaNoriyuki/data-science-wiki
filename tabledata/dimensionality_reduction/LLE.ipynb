# Locally Linear Embedding

```python
from sklearn.manifold import LocallyLinearEmbedding

import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
```

## データの用意

```python
dataset = load_wine()
```

```python
dataset.data
```

```python
label = dataset.target
```

## LLEの実行

```python
lle_5 = LocallyLinearEmbedding(n_neighbors=5, n_components=2)
X_lle_5 = lle_5.fit_transform(dataset.data)


lle_10 = LocallyLinearEmbedding(n_neighbors=10, n_components=2)
X_lle_10 = lle_10.fit_transform(dataset.data)


lle_20 = LocallyLinearEmbedding(n_neighbors=20, n_components=2)
X_lle_20 = lle_20.fit_transform(dataset.data)
```

## 結果の可視化

```python
plt.scatter(X_lle_5[:,0],X_lle_5[:,1], c=label)
plt.xlabel("isomap_n5-1")
plt.ylabel("isomap_n5-2")
```

```python
plt.scatter(X_lle_10[:,0],X_lle_10[:,1], c=label)
plt.xlabel("lle_n10-1")
plt.ylabel("lle_n10-2")
```

```python
plt.scatter(X_lle_20[:,0],X_lle_20[:,1], c=label)
plt.xlabel("lle_n20-1")
plt.ylabel("lle_n20-2")
```
