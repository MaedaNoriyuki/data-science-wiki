# Isomap

```python
from sklearn.manifold import Isomap

import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
```

## データの用意

```python
dataset = load_wine()
```

```python
dataset.data
```

```python
label = dataset.target
```

## Isomapの実行

```python
# 近傍5個を考慮
isomap_5 = Isomap(n_neighbors=5, n_components=2)
X_isomap_5 = isomap_5.fit_transform(dataset.data)

# 近傍10個を考慮
isomap_10 = Isomap(n_neighbors=10, n_components=2)
X_isomap_10 = isomap_10.fit_transform(dataset.data)

# 近傍20個を考慮
isomap_20 = Isomap(n_neighbors=20, n_components=2)
X_isomap_20 = isomap_20.fit_transform(dataset.data)
```

## 結果の可視化

```python
plt.scatter(X_isomap_5[:,0],X_isomap_5[:,1], c=label)
plt.xlabel("isomap_n5-1")
plt.ylabel("isomap_n5-2")
```

```python
plt.scatter(X_isomap_10[:,0],X_isomap_10[:,1], c=label)
plt.xlabel("isomap_n10-1")
plt.ylabel("isomap_n10-2")
```

```python
plt.scatter(X_isomap_20[:,0],X_isomap_20[:,1], c=label)
plt.xlabel("isomap_n20-1")
plt.ylabel("isomap_n20-2")
```
