```python
import umap

import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
```

## データの用意

```python
dataset = load_wine()
```

```python
dataset.data
```

```python
label = dataset.target
```

## UMAPの実行

```python
embedding = umap.UMAP().fit_transform(dataset.data)
```

## 結果の可視化

```python
plt.scatter(embedding[:,0],embedding[:,1], c=label)
plt.xlabel("UMAP-1")
plt.ylabel("UMAP-2")
```
