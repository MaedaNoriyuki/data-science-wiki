# DBSCAN

DBSCANは密度ベースのクラスタリング手法でクラスタ数を事前に設定することなく，異なる密度領域を別のクラスターとして判定します．

代表的なパラメータ
- eps
    - データポイントが他のポイントと"近い"と見なされる最大距離。この距離内にあるポイントが密な領域を形成します。
- min_samples
    - コアポイントであると見なすために必要な、eps距離内の最小ポイント数。この数以上の近傍ポイントを持つ点は、クラスタを形成する核となります。

```python
from sklearn import cluster, datasets
import matplotlib.pyplot as plt
```

```python
data = datasets.make_moons(n_samples=1000, noise=.05)[0]
```

### DBSCAN

```python
dbscan_data = cluster.DBSCAN(eps=0.1, min_samples=5, metric='euclidean').fit_predict(data)
```

```python
plt.scatter(data[:,0], data[:,1], c=dbscan_data, cmap='plasma', s=20)
plt.title('DBSCAN Clustering')
plt.show()
```
