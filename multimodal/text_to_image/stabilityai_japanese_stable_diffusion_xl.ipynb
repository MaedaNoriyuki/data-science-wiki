```python
from diffusers import DiffusionPipeline
import torch
```

```python
pipeline = DiffusionPipeline.from_pretrained("stabilityai/japanese-stable-diffusion-xl", trust_remote_code=True)
pipeline.to("cuda")
```

```python
prompt = "柴犬、カラフルアート"

image = pipeline(prompt=prompt).images[0]
image
```
