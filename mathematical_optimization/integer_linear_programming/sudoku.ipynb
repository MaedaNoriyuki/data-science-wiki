```python
import pulp 
import numpy as np
```

```python
problem = pulp.LpProblem('数独',sense=pulp.LpMinimize)

var = []

for i in range(9):
  sublist = []
  for j in range(9):
    subsublist = []
    for k in range(9):
      subsublist.append(pulp.LpVariable(name= f'x{i}_{j}_{k}',cat = pulp.LpBinary)) #変数の名前は全ての異なる名前にする必要がある
    sublist.append(subsublist)
  var.append(sublist)

print(var)
```

```python
#制約条件


#一つのマス目に対する変数の制約
for i in range(9):
  for j in range(9):
    problem += pulp.lpSum(var[i][j]) == 1

#列ごとの同じ値の変数の制約
for k in range(9):
  a = [row[k] for row in var] #列の値
  for l in range(9):
    b = [row[l] for row in a]  #数値の値
    problem += pulp.lpSum(b) == 1

#行ごとの同じ値の変数の制約
for m in range(9):
  c = var[m]                  #行の値
  for n in range(9):
    d = [row[n] for row in c] #数値の値
    problem += pulp.lpSum(d) == 1


number = [1,2,3,4,5,6,7,8,9]

#行ごとの和の制限
for i in range(9):
  problem += pulp.lpDot(var[i][0],number) + pulp.lpDot(var[i][1],number) + pulp.lpDot(var[i][2],number) + pulp.lpDot(var[i][3],number) + pulp.lpDot(var[i][4],number) + pulp.lpDot(var[i][5],number) + pulp.lpDot(var[i][6],number) + pulp.lpDot(var[i][7],number) + pulp.lpDot(var[i][8],number) == 45


#列ごとの和の制限
for j in range(9):
  problem += pulp.lpDot(var[0][j],number) + pulp.lpDot(var[1][j],number) + pulp.lpDot(var[2][j],number) + pulp.lpDot(var[3][j],number) + pulp.lpDot(var[4][j],number) + pulp.lpDot(var[5][j],number) + pulp.lpDot(var[6][j],number) + pulp.lpDot(var[7][j],number) + pulp.lpDot(var[8][j],number) == 45


#3×3のブロックの制約
for k in range(3):
  for j in range(3):
    for i in range(9):
      problem += var[3*k][3*j][i] + var[3*k][3*j+1][i] + var[3*k][3*j+2][i] +var[3*k+1][3*j][i] + var[3*k+1][3*j+1][i] + var[3*k+1][3*j+2][i] +var[3*k+2][3*j][i] + var[3*k+2][3*j+1][i] + var[3*k+2][3*j+2][i] == 1



#目的関数は特にないので設定しない

#定式化を見る
problem 
```

```python
#解を求める
result = problem.solve()

#答えをリストに格納
res = []

for i in range(9):
  subres = []
  for j in range(9):
    subsubres = []
    for k in range(9):
      subsubres.append(pulp.value(var[i][j][k])) #変数の名前は全ての異なる名前にする必要がある
    subres.append(subsubres)
  res.append(subres)

#リストをarrauに変換
res_1 = np.array(res)
#1の値になっているもののインデックスを抽出
res_index = np.where(res_1 == 1)

#インデックスの「数値」に対応する箇所を取り出し1を加える
sudoku_result = np.reshape(res_index[2] + 1,(9,9))
display(sudoku_result)
```
