```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

from google.colab import drive 
drive.mount('/content/drive')
```

## パスの設定

```python
# 以下のサイトからフォントをダウンロードする
# https://moji.or.jp/ipafont/ipa00303/
IPA_path = '/content/drive/MyDrive/Colab Notebooks/ipagp.ttf'
file_name = "sample.png"
```

## 入力テキスト

```python
text = '量子計算 VQE QAOA QWOA ボソンサンプリング 量子自然言語処理 量子回路学習　量子回路ボルンマシン　量子ボルツマンマシン　テンソルネットワーク　\
量子強化学習 量子自然言語処理 量子オートエンコーダ 量子GAN 量子主成分分析 量子SVM \
量子化学計算 量子線形回帰 　ガウシアンボソンサンプリング 量子ニューラルネットワーク \
デジタルアニーラ QCNN 量子アニーリング 量子断熱計算 QGRNN QuantumTransferLearning グローバーのアルゴリズム \
量子モンテカルロ法 量子ウォーク pennylane Qiskit TensorFlowQuantum Cirq blueqat Qulacs StrawberryFeilds 量子勾配降下法 \
Yao Ocean 変分量子アルゴリズム バレンプラトー 量子リザバーコンピューティング 量子古典ハイブリッド計算 QUBO イジングハミルトニアン \
量子位相推定 量子振幅推定　量子フーリエ変換 光連続量量子計算 量子ゲート'
```

## wordcloudの出力

```python
wordcloud = WordCloud(max_font_size=45,
                      background_color='white',
                      font_path = IPA_path,
                      max_words=2000,
                      width=800,
                      height=400).generate(text)

plt.figure(figsize=(20,30))
plt.imshow(wordcloud)
plt.axis("off")

wordcloud.to_file('wordcloud.png')
```

## シルエットを考慮したwordcloud

```python
mask = np.array(Image.open("alice_mask.png"))

# マスク画像の表示
plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud = WordCloud(max_font_size=40,
                      background_color='white',
                      max_words=2000,
                      mask=mask,
                      font_path = IPA_path,
                      contour_color='white').generate(text)

plt.figure(figsize=(15,12))
plt.imshow(wordcloud)
plt.axis("off")
#plt.savefig("word_cloud.png")
```
